# Data cleaning

The aim of this notebook is to :
* Merge all ads of all political parties by adding a column parti
* Clean All the columns :
    * Make index id (int)
    * Transform date into datetime
    * 

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from glob import glob

DATA_PARTI_PATH = '../data/ad_by_partis/'
GENERATED_PATH = '../generated/'
```

```python
# Merge csv files
all_ads = pd.DataFrame()
for filename in glob(DATA_PARTI_PATH+'*.csv'):
    # Get the name of the parti
    parti = filename.replace(DATA_PARTI_PATH, '')
    parti = parti.replace('.csv','')
    
    ads = pd.read_csv(filename, index_col=0)
    ads['parti'] = parti
    all_ads = all_ads.append(ads, sort=False)
```

```python
# Clean date
all_ads['date'] = all_ads['date'].str.strip('D√©but de diffusion le ')
all_ads['date'] = all_ads['date'].str.replace('d√©c', 'dec')
all_ads['date'] = all_ads['date'].str.replace('ao√ª', 'aug')
all_ads['date'] = all_ads['date'].str.replace('juil', 'jul')
all_ads['date'] = all_ads['date'].str.replace('mai', 'may')
all_ads['date'] = all_ads['date'].str.replace('avr', 'apr')
all_ads['date'] = all_ads['date'].str.replace('f√©v', 'feb')
all_ads['date'] = all_ads['date'].str.split('-').apply(lambda x : x[0])
all_ads['date'] = pd.to_datetime(all_ads['date'])
```

```python
# Set index to id
all_ads['id'] = all_ads['id'].str.strip('ID\xa0: ').astype('int')
all_ads = all_ads.set_index('id', drop=True)
```

We will only get the ads from 2019 :

```python
all_ads['year'] = pd.DatetimeIndex(all_ads['date']).year
all_ads = all_ads[all_ads['year']==2019]
all_ads = all_ads.drop(columns='year')
all_ads
```

```python
all_ads.to_csv(GENERATED_PATH+'clean_ads.csv')
```
